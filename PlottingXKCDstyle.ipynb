# Plotting, XKCD style.

![](https://imgs.xkcd.com/comics/self_description.png)

We can even plot in different styles, such as XKCD, made possible by [Jake VanderPlas](http://jakevdp.github.io/blog/2013/07/10/XKCD-plots-in-matplotlib/).

**Note**: In order to do this, first <a href="http://antiyawn.com/uploads/humorsans.html">install Humor Sans</a> and then restart your matplotlib cache with `matplotlib.font_manager._rebuild()`.

```python
# Import necessary packages
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import math
```

```python
# Rebuild your fonts cache
matplotlib.font_manager._rebuild()
```

```python
plt.xkcd()
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(211)

years = np.linspace(1975, 2013)
pct = 2 + 98. / (1 + np.exp(0.6 * (2008 - years)))
ax.plot(years, pct)

ax.set_xlim(1976, 2013)
ax.set_ylim(0, 100)
ax.yaxis.set_major_formatter(plt.FormatStrFormatter('%i%%'))

ax.text(1977, 67,
        ("Percentage of the US Population\n"
         "carrying cameras everywhere they go,\n"
         "every waking moment of their lives:"),
        size=16)

ax.set_xlabel(("In the last few years, with very little fanfare,\n"
               "We've conclusively settled the questions of\n"
               "flying saucers, lake monsters, ghosts, and bigfoot."),
              size=16)
plt.show()
```

```python
x = np.linspace(0, 10)
y1 = x * np.sin(x)
y2 = x * np.cos(x)

plt.fill(x, y1, 'red', alpha=0.4)
plt.fill(x, y2, 'blue', alpha=0.4)
plt.xlabel('x axis yo!')
plt.ylabel("I don't even know")
plt.show()
```
